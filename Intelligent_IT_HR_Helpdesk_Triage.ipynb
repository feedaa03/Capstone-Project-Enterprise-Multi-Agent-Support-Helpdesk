### рџ“љ CELL 2: LOCAL VECTOR DATABASE INGESTION FROM /CONTENT/ FILE
**Purpose & Design:** This block orchestrates the active **Retrieval-Augmented Generation (RAG)** ingestion layer. The architecture physically reads the compiled raw data from `/content/wiki_data.txt`, tokenizes and processes the corporate context into distinct semantic snippets (Chunks), and embeds them into low-dimensional vector representations utilizing the `all-MiniLM-L6-v2` sentence-transformer.

These dense matrices are indexed inside a transient local instance of **ChromaDB**, exposing a deterministic `retrieve_knowledge_context` lookup loop that prevents LLM hallucination during informational employee auditing.

```python
## =====================================================================
# CELL 2: LOCAL VECTOR DATABASE INGESTION FROM /CONTENT/ FILE
# =====================================================================
import chromadb
from sentence_transformers import SentenceTransformer

# 1. ШЁЩ†Ш§ШЎ Щ€Щ‚Ш±Ш§ШЎШ© Ш§Щ„ШЁЩЉШ§Щ†Ш§ШЄ Щ…ШЁШ§ШґШ±Ш© Щ…Щ† Ш§Щ„Щ…Щ„ЩЃ Ш§Щ„Щ…Щ€Ш¬Щ€ШЇ ЩЃЩЉ Щ…Ш¬Щ„ШЇ Ш§Щ„ЩѓЩ€Щ†ШЄЩ†ШЄ
with open("/content/wiki_data.txt", "r", encoding="utf-8") as f:
    physical_file_content = f.read()

# ШЄЩ‚ШіЩЉЩ… Ш§Щ„Щ†Шµ ШҐЩ„Щ‰ ЩЃЩ‚Ш±Ш§ШЄ (Chunks) ШЁЩ†Ш§ШЎЩ‹ Ш№Щ„Щ‰ Ш§Щ„Щ…Щ€Ш§Ш¶ЩЉШ№
wiki_chunks = [chunk.strip() for chunk in physical_file_content.split("\n\n") if chunk.strip()]

# 2. ШЄЩ‡ЩЉШ¦Ш© Щ€ШЄШ¬Щ‡ЩЉШІ Щ‚Ш§Ш№ШЇШ© ШЁЩЉШ§Щ†Ш§ШЄ Ш§Щ„Щ…ШЄШ¬Щ‡Ш§ШЄ ChromaDB Щ…Ш­Щ„ЩЉШ§Щ‹
chroma_client = chromadb.Client()
collection = chroma_client.get_or_create_collection(name="helpdesk_knowledge_base")

# Ш§ШіШЄШЇШ№Ш§ШЎ Щ…Щ€ШЇЩЉЩ„ Ш§Щ„ШЄШ¶Щ…ЩЉЩ† Ш§Щ„Ш±ЩЉШ§Ш¶ЩЉ Ш§Щ„Ш®ЩЃЩЉЩЃ Щ€Ш§Щ„Щ…ШµШ±Ш­ ШЁЩ‡
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# ШґШ­Щ† Щ€ШЄЩѓШґЩЉЩЃ Ш§Щ„ШЁЩЉШ§Щ†Ш§ШЄ ШЇШ§Ш®Щ„ Щ‚Ш§Ш№ШЇШ© Ш§Щ„ШЁЩЉШ§Щ†Ш§ШЄ
for i, chunk in enumerate(wiki_chunks):
    vector_embedding = embedding_model.encode(chunk).tolist()
    collection.add(
        ids=[f"wiki_node_{i}"],
        embeddings=[vector_embedding],
        documents=[chunk]
    )

def retrieve_knowledge_context(user_query: str) -> list:
    """ШЇШ§Щ„Ш© Ш§Щ„ШЁШ­Ш« Ш§Щ„Ш§ШіШЄШ±Ш¬Ш§Ш№ЩЉ Ш§Щ„ШЇЩ„Ш§Щ„ЩЉ (RAG Semantic Retrieval)"""
    query_vector = embedding_model.encode(user_query).tolist()
    search_results = collection.query(query_embeddings=[query_vector], n_results=1)

    if search_results and search_results['documents']:
        return search_results['documents'][0]
    return ["No verified corporate context found for this inquiry."]

print("вњ… Cell 2 Complete: Local ChromaDB successfully populated by reading directly from /content/wiki_data.txt!")
```
